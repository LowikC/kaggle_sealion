```python
def random_flip(x, y):
    flips = np.random.randint(0, 2, (3,))
    print(flips)
    if flips[0]:
        x = np.rot90(x)
        y = np.rot90(y)
    if flips[1]:
        x = np.flipud(x)
        y = np.flipud(y)
    if flips[2]:
        x = np.fliplr(x)
        y = np.fliplr(y)
    return x, y
```
